# BANK DATASET

```python
bd=pd.read_csv("Bank.csv")
```

```python
bd.head()
```

```python
bd.isnull().any()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
bd["deposit"] = le.fit_transform(bd["deposit"])
```

```python
bd.head()
```

```python
x = bd.iloc[:,0:16].values
y = bd.iloc[:,16].values
```

```python
y
```

```python
x
```

```python
x.shape
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer([("oh",OneHotEncoder(),[1,2,3,4,6,7,8,10,15])],remainder = "passthrough")  #converts data to binary format
x = ct.fit_transform(x)
```

```python
x
```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 10000)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
x_test
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(x_train,y_train)
```

```python
ypred = logreg.predict(x_test)
```

```python
ypred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(ypred,y_test)
```

```python
accuracy
```
